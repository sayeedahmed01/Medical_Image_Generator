```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Set up the device
torch_device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the Stable Diffusion pipeline
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(torch_device)

# Set up generation parameters
prompt = ["generate an image of an acne type of rash on the cheek area on a fair skin"]
height = 512  # default height of Stable Diffusion
width = 512  # default width of Stable Diffusion
num_inference_steps = 25  # Number of denoising steps
guidance_scale = 7.5  # Scale for classifier-free guidance
generator = torch.Generator(device=torch_device).manual_seed(0)  # Set seed for reproducibility

# Generate the image
with torch.autocast(torch_device):
    image = pipe(
        prompt,
        height=height,
        width=width,
        num_inference_steps=num_inference_steps,
        guidance_scale=guidance_scale,
        generator=generator
    ).images[0]

# Display or save the image
image.show()  # To display the image
# image.save("generated_image.png")  # To save the image
```
