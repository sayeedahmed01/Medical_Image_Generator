```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
import pandas as pd
import random
import matplotlib.pyplot as plt
import os
import numpy as np
from sklearn.model_selection import train_test_split
from tqdm import tqdm

import torch.nn.functional as F
from transformers import get_linear_schedule_with_warmup
```

```python
data_dir = "/content/drive/MyDrive/CLIP_training_data"
metadata_path = os.path.join(data_dir, "metadata.csv")
```

```python
# Load dataset
df = pd.read_csv(metadata_path)
```

```python
class CustomDataset(Dataset):
    def __init__(self, dataframe, processor, data_dir):
        self.dataframe = dataframe
        self.processor = processor
        self.data_dir = data_dir

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_path = os.path.join(self.data_dir, row['image_path'])
        image = Image.open(image_path).convert('RGB')
        text = row['text']

        inputs = self.processor(text=[text], images=image, return_tensors="pt", padding='max_length', truncation=True)

        return {
            'input_ids': inputs['input_ids'].squeeze(),
            'attention_mask': inputs['attention_mask'].squeeze(),
            'pixel_values': inputs['pixel_values'].squeeze()
        }
```

```python
# Load pre-trained CLIP model and processor
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Set up optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
```

```python
# Single image processing and visualization
image_path = os.path.join(data_dir, df.iloc[0]['image_path'])
image = Image.open(image_path).convert('RGB')
inputs = processor(text=["Test"], images=image, return_tensors="pt", padding='max_length', truncation=True)

# Get pixel values and visualize
image_processed = inputs['pixel_values'].squeeze().permute(1, 2, 0).numpy()
image_processed = (image_processed - image_processed.min()) / (image_processed.max() - image_processed.min())
plt.imshow(image_processed)
plt.show()

# Print min and max values for debugging
print(f"Sanity Check - Raw Min: {inputs['pixel_values'].min():.4f}, Raw Max: {inputs['pixel_values'].max():.4f}")
print(f"Sanity Check - Min After Normalization: {image_processed.min():.4f}, Max After Normalization: {image_processed.max():.4f}")
```

```python
# Experiment 1
```

```python
dataset = CustomDataset(df, processor, data_dir)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
import matplotlib.pyplot as plt
import random
import torch

# Sample and display a few images
num_samples = 5
sampled_indices = random.sample(range(len(dataset)), num_samples)

fig, axs = plt.subplots(num_samples, 1, figsize=(10, 5*num_samples))
fig.tight_layout(pad=5.0)

for i, idx in enumerate(sampled_indices):
    sample = dataset[idx]

    # Access the 'pixel_values' tensor using the key 'pixel_values'
    image = sample['pixel_values'].cpu()  # Ensure tensor is on CPU
    image = image.permute(1, 2, 0).numpy()  # Convert to numpy array and change to [H, W, C] format

    # Convert to [0, 1] range for display, if necessary (this might not be necessary if processor does this)
    image = (image - image.min()) / (image.max() - image.min())

    # Display the image
    axs[i].imshow(image)
    axs[i].set_title(f"Image {i+1}")
    axs[i].axis('off')

    # Print the associated text
    caption = df.iloc[idx]['text']
    axs[i].set_xlabel(f"Caption: {caption}", wrap=True)

    # Print min and max values for debugging
    print(f"Image {i+1} - Raw Min: {sample['pixel_values'].min():.4f}, Raw Max: {sample['pixel_values'].max():.4f}")
    print(f"Image {i+1} - Min After Normalization: {image.min():.4f}, Max After Normalization: {image.max():.4f}")

plt.show()

# Print captions separately for better readability
for i, idx in enumerate(sampled_indices):
    print(f"Image {i+1} Caption: {df.iloc[idx]['text']}")
```

```python
# Training loop
num_epochs = 3
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=100, num_training_steps=len(dataloader) * num_epochs)


train_losses = []
val_losses = []
val_accuracies = []

for epoch in range(num_epochs):
    total_loss = 0  # Accumulate loss for the epoch
    for i, batch in enumerate(dataloader):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        pixel_values = batch['pixel_values'].to(device)

        # Create labels based on the actual batch size
        labels = torch.arange(input_ids.shape[0], device=device)

        # Calculate the outputs
        outputs = model(input_ids=input_ids, attention_mask=attention_mask,
                        pixel_values=pixel_values)

        # Extract the logits for image and text modalities
        logits_per_image = outputs.logits_per_image
        logits_per_text = outputs.logits_per_text

        # Calculate the contrastive loss
        loss = (F.cross_entropy(logits_per_image, labels) +
                F.cross_entropy(logits_per_text, labels)) / 2

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()

        # Print progress every 10 batches
        if (i + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Batch {i+1}, Loss: {loss.item()}")

    # Calculate and print average epoch loss
    avg_epoch_loss = total_loss / len(dataloader)
    print(f"Epoch {epoch+1}/{num_epochs}, Average Loss: {avg_epoch_loss}")

    # Save checkpoint after each epoch
    model.save_pretrained(f"clip_finetuned_checkpoint_epoch_{epoch+1}")
    processor.save_pretrained(f"clip_finetuned_checkpoint_epoch_{epoch+1}")

# Save the final fine-tuned model
model.save_pretrained("clip_finetuned")
processor.save_pretrained("clip_finetuned")
```

```python
# Train-Validation Split
train_df, val_df = train_test_split(df, test_size=0.1, random_state=42)

print(f"Full dataset size: {len(df)}")
print(f"Training set size: {len(train_df)}")
print(f"Validation set size: {len(val_df)}")
```

```python
# Create dataset and dataloader
train_dataset = CustomDataset(train_df, processor, data_dir)
val_dataset = CustomDataset(val_df, processor, data_dir)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
import matplotlib.pyplot as plt
import random
import torch

# Sample and display a few images
num_samples = 5
sampled_indices = random.sample(range(len(train_dataset)), num_samples)

fig, axs = plt.subplots(num_samples, 1, figsize=(10, 5*num_samples))
fig.tight_layout(pad=5.0)

for i, idx in enumerate(sampled_indices):
    sample = train_dataset[idx]

    # Access the 'pixel_values' tensor using the key 'pixel_values'
    image = sample['pixel_values'].cpu()  # Ensure tensor is on CPU
    image = image.permute(1, 2, 0).numpy()  # Convert to numpy array and change to [H, W, C] format

    # Convert to [0, 1] range for display, if necessary (this might not be necessary if processor does this)
    image = (image - image.min()) / (image.max() - image.min())

    # Display the image
    axs[i].imshow(image)
    axs[i].set_title(f"Image {i+1}")
    axs[i].axis('off')

    # Print the associated text
    caption = df.iloc[idx]['text']
    axs[i].set_xlabel(f"Caption: {caption}", wrap=True)

    # Print min and max values for debugging
    print(f"Image {i+1} - Raw Min: {sample['pixel_values'].min():.4f}, Raw Max: {sample['pixel_values'].max():.4f}")
    print(f"Image {i+1} - Min After Normalization: {image.min():.4f}, Max After Normalization: {image.max():.4f}")

plt.show()

# Print captions separately for better readability
for i, idx in enumerate(sampled_indices):
    print(f"Image {i+1} Caption: {df.iloc[idx]['text']}")
```

```python
num_epochs = 15
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def contrastive_loss(logits):
    labels = torch.arange(len(logits), device=logits.device)
    return (F.cross_entropy(logits, labels) + F.cross_entropy(logits.T, labels)) / 2

def save_model(model, processor, path):
    model.save_pretrained(path)
    processor.save_pretrained(path)

# Create a directory to save models
os.makedirs('model_checkpoints', exist_ok=True)

best_val_loss = float('inf')
save_interval = 5  # Save every 5 epochs

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{num_epochs}")

    for batch in progress_bar:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        pixel_values = batch['pixel_values'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, pixel_values=pixel_values)
        loss = contrastive_loss(outputs.logits_per_image)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        progress_bar.set_postfix({'training_loss': f'{loss.item():.3f}'})

    avg_train_loss = total_loss / len(train_dataloader)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {avg_train_loss:.4f}")

    # Validation loop
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for batch in val_dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            pixel_values = batch['pixel_values'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, pixel_values=pixel_values)
            loss = contrastive_loss(outputs.logits_per_image)
            val_loss += loss.item()

    avg_val_loss = val_loss / len(val_dataloader)
    print(f"Epoch {epoch+1}/{num_epochs}, Validation Loss: {avg_val_loss:.4f}")

    # Save the best model
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        save_model(model, processor, 'model_checkpoints/best_model')
        print(f"New best model saved with validation loss: {best_val_loss:.4f}")

    # Save model at regular intervals
    if (epoch + 1) % save_interval == 0:
        save_model(model, processor, f'model_checkpoints/model_epoch_{epoch+1}')
        print(f"Model saved at epoch {epoch+1}")

# Save the final model
save_model(model, processor, 'model_checkpoints/final_model')
print("Training completed. Final model saved.")
```

```python
def run_inference(model, processor, image_path, text_options, actual_prompt):
    # Load and preprocess the image
    image = Image.open(image_path).convert('RGB')
    inputs = processor(images=image, text=text_options, return_tensors="pt", padding=True)

    # Move inputs to the same device as the model
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    # Run inference
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the logits
    logits_per_image = outputs.logits_per_image
    probs = logits_per_image.softmax(dim=1)

    # Display the image
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.axis('off')
    plt.title('Input Image')
    plt.show()

    # Print actual prompt
    print(f"Actual Prompt: {actual_prompt}")

    # Print probabilities
    print("\nPrediction Probabilities:")
    for text, prob in zip(text_options, probs[0]):
        print(f"{text}: {prob.item():.4f}")

# After training loop
print("Running inference on validation images...")

# Load the best model
best_model = CLIPModel.from_pretrained('model_checkpoints/best_model')
best_model.to(device)
best_model.eval()

# Choose some validation images and text options
num_samples = 5
sampled_indices = random.sample(range(len(val_dataset)), num_samples)

for idx in sampled_indices:
    sample = val_dataset[idx]
    image_path = os.path.join(data_dir, val_df.iloc[idx]['image_path'])
    actual_prompt = val_df.iloc[idx]['text']

    # Generate some text options (including the actual prompt)
    text_options = [actual_prompt] + random.sample(list(val_df['text']), 3)

    print(f"\nInference for image: {image_path}")

    run_inference(best_model, processor, image_path, text_options, actual_prompt)
```
